```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
okavango_rainfall='results/okavango_rainfall_2013-12_to_2021-05.csv'
upstream_rainfall='results/upstream_rainfall_2013-12_to_2021-05.csv'
water_extent='results/water_extent_2013-12_to_2021-05.csv'
```

```python
df=pd.read_csv(okavango_rainfall, index_col='time', parse_dates=True)[0:-1]
df.index = df.index.to_period('M')
upr=pd.read_csv(upstream_rainfall, index_col='time',parse_dates=True)[0:-1]
upr.index = upr.index.to_period('M')
we=pd.read_csv(water_extent, index_col='time',parse_dates=True)
we.index = we.index.to_period('M')
```

```python
upr.index.values[1]
```

```python
type(we.index.values[0])
```

```python
df['upstream_rain']=upr
df['water_extent']=we
```

```python
df = df.rename({'cumulative 3-month rainfall (mm)':'okavango_rain'},axis=1)
```

```python
df.plot(figsize=(18,7), linestyle='dashed', marker='o');
plt.ylabel('Open water extent (km2) / cumulative rainfall (mm)')
```
